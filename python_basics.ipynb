Python is high-level language - meant to be read by the ppl

It's an interpreted language meaning it's not compiled into machine language, and it's used in interactive fashion. that's useful for tasks that require a lot of investigation rather than planning.

It's dynamically typed like Javascript. there's no types. 

A variable can store any time.


```python
def add_numbers(x, y, z=None):
    if z != None:
        return x + y + z
    else:
        return x + y

add_numbers(2, 3, 1)
```

```python
x = 1
print( type(x))

x = "sfdda"
print( type(x))

x = False
print( type(x))

x = 1.0
print( type(x))

x = add_numbers
print( type(x))
```

```python
#Dictionaries aka Maps
d = {'key':'value', 'john_smith': 'js@mail.com'}
d
```

```python
print('Iterate keys')
for k in d:
    print(k);

print('\nIterate values')
for v in d.values():
    print(v);
    
# unpacking
print('\nIterate key and values')
for k, v in d.items():
    print('{} -> {}'.format(k, v))
```

Python Dates and Times

```python
import datetime as dt
import time as tm

tm.time()
```

```python
dtnow = dt.datetime.fromtimestamp(tm.time())
dtnow
```

```python
dtnow.year, dtnow.month, dtnow.day, dtnow.hour, dtnow.minute, dtnow.second
```

```python
delta = dt.timedelta(days=100)
delta
```

```python
today = dt.date.today()
today - delta
```

```python
today > today - delta
```
