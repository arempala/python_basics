```python
#Dictionaries aka Maps
d = {'key':'value', 'john_smith': 'js@mail.com'}
d
```

```python
print('Iterate keys')
for k in d:
    print(k);

print('\nIterate values')
for v in d.values():
    print(v);
    
# unpacking
print('\nIterate key and values')
for k, v in d.items():
    print('{} -> {}'.format(k, v))
```

Functional programming

map()

```python
store1 = [10.00, 11.00, 12.34, 2.34]
store2 = [9.00, 11.00, 12.34, 2.01]
cheapest = map(min, store1, store2)

for i in cheapest:
    print(i)
```

```python
#write a function and apply it using map() to get a list of all faculty titles and last names

people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(person):
    return person.split(' ')[0] + ' ' + person.split(' ')[2]

list(map(split_title_and_name, people))
```

Advanced Python Lambda and List Comprehension

my_function = lambda a, b, c : a + b
my_function(1,2,3)

```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

list(map(lambda p : p.split(' ')[0] + ' ' + p.split(' ')[2], people))
```

List comprehension

```python
my_list = []
for number in range(0,1000):
    if number % 2 == 0:
        my_list.append(number)
        
# the above can be re-written w/ list comprehension
my_list = [number for number in range(0,1000) if number % 2 == 0]

my_list[:4]
```

```python
lst = [i*j for i in range(0,10) for j in range(0,10)]
lst[10:15]
```
